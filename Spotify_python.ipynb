```python
#Bibliotheken importieren
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Datensatz lesen
data = pd.read_csv(r'C:\Users\shkelqim.haliti\Desktop\spotify_top_charts_22.csv')
#Datensatz anzeigen lassen
data.head(10)
```

```python
#Matrixform anzeigen lassen
data.shape
```

```python
#Schauen ob Datensätze fehlen
for i in data.columns:
    daten_fehlen = np.mean(data[i].isnull())
    print('{} - {}%'.format(i,daten_fehlen))
```

```python
#Datentypen anzeigen lassen und schauen ob eine Null im Datensatz ist
data.info()
```

```python
#Datentyp ändern und Spalte umbenennen
data['URL'] = data['uri'].astype(str)
```

```python
#Irrelevante Spalten löschen
columns = ['uri','time_signature']
data.drop(columns, inplace=True, axis=1)
```

```python
#Redondante Datensätze löschen (hier nicht nötig, aber ich möchte ein Beispiel zeigen)
#data['artist_names'].drop_duplicates().sort_values(ascending=False)
```

```python
#Statistische Daten ausgeben lassen
data.describe()
```

```python
#Welche Künstler waren am häufigsten in den Charts?
data['artist_names'].value_counts().head(15)
```

```python
#Mit einem Tortendiagramm visualisieren
data['artist_names'].value_counts().head(15).plot(kind='pie', figsize=(6,6))
```

```python
#Wie viele Wochen bleiben die Songs von The Weeknd in den Charts?
#Zuerst einen Untersatensatz erstellen
TheWeeknd = data.loc[data['artist_names'] == 'The Weeknd']
TheWeeknd
```

```python
#Mit einem Balkendiagramm visualisieren
plt.barh(TheWeeknd.track_name, TheWeeknd.weeks_on_chart, color='green') 
plt.title("Balkendiagramm")
plt.xticks(rotation='vertical')
plt.ylabel("Song")
plt.xlabel("Anzahl der Wochen")
plt.legend(["Anzahl der Wochen in den Charts"])
plt.show()
```

```python
#Wie viele Minuten dauert durchschnittlich ein Song in den Charts?
data.duration_ms.describe()/60000 #Wir müssen durch 60000 teilen, da die Angaben in Millisekunden sind
```

```python
#Mit einem Boxplot visualisieren
plt.boxplot(data.duration_ms/60000, vert = False)
plt.title("Boxplot")
plt.xlabel("Zeit in Sekunden")
plt.legend(["Dauer eines Songs in den Charts"])
plt.show()
```

```python
#Wie lange dauern die Songs von Drake in den Charts?
#Zuerst einen Unterdatensatz erstellen
Drake = data.loc[data['artist_names'] == 'Drake']
Drake
```

```python
#Mit einem Graphen visualisieren
plt.plot(Drake.track_name, Drake.duration_ms/60000, color="green")
plt.title("Graph")
plt.xticks(rotation='vertical')
plt.xlabel("Song")
plt.ylabel("Dauer in Minuten")
plt.legend(["Dauer der Songs"])
plt.show()
```

```python
#Wie sieht bei den Charts songs von Bad Bunny die danceability, energy und tempo aus?
#Zuerst einen Unterdatensatz erstellen
BadBunny = data.loc[data['artist_names'] == 'Bad Bunny']
BadBunnyKriterien = BadBunny[["danceability","energy", "tempo"]]
```

```python
#Histogramme mit einer Schleife visualisieren
for i in BadBunnyKriterien.columns:
    plt.hist(BadBunnyKriterien[i])
    plt.title(i)
    plt.xlabel("Wert")
    plt.ylabel("Song")
    plt.legend([i])
    plt.show()
```

```python
#Korrelation anzeigen
data.corr()
```

```python
#Korrelation visualisieren
sns.heatmap(data.corr(), annot=True)
plt.title('Korrelationsmatrix')
```

```python
#Korrelationspaare anzeigen lassen und nach Korrelation sortieren
corr_mat = data.corr()
corr_paare = corr_mat.unstack()
paare_sortiert = corr_paare.sort_values()
paare_sortiert.head(10)
```

```python
#Mit einem Sreudiagramm die im Model höchste Korrelation zwischen energy und acousticness visualisieren
plt.scatter(data.energy, data.acousticness, color="blue")
plt.title("Streudiagramm")
plt.xlabel("energy")
plt.ylabel("acousticness")
plt.show()
```

```python
#Wir konvertieren die Künstlernamen nummerisch, damit wir auch dort die Korrelation überprüfen und eventuell Zusammenhänge erkennen
df_nummeriert = data.copy()

for col_name in df_nummeriert.columns:
    if(df_nummeriert[col_name].dtype == 'object'):
        df_nummeriert[col_name] = df_nummeriert[col_name].astype('category')
        df_nummeriert[col_name] = df_nummeriert[col_name].cat.codes

df_nummeriert
```

```python
#Korrelation des neuen Datensatzes anzeigen lassen
print(df_nummeriert.corr())
```

```python
#Korrelation visualisieren
sns.heatmap(df_nummeriert.corr(), annot=True)
plt.title('Korrelationsmatrix inklusive objects')
```

```python
#Korrelationspaare anzeigen lassen und nach größter Korrelation sortieren
corr_mat2 = df_nummeriert.corr()
corr_paare2 = corr_mat2.unstack()
paare_sortiert2 = corr_paare2.sort_values()
paare_sortiert2.head(10)
```
